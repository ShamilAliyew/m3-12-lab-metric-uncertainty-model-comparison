## Setup

Run the cell below to import packages, generate the dataset, and split it into train/test sets. The synthetic dataset is designed so that simple models perform reasonably well and the gap between them is small — exactly the scenario where statistical comparison matters.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    classification_report,
    f1_score,
    precision_score,
    recall_score,
)
from scipy import stats

SEED = 42
np.random.seed(SEED)

# Generate a binary classification dataset
X, y = make_classification(
    n_samples=1200,
    n_features=15,
    n_informative=8,
    n_redundant=3,
    n_classes=2,
    flip_y=0.08,
    random_state=SEED,
)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=SEED
)

print(f"Training set: {X_train.shape[0]} samples")
print(f"Test set:     {X_test.shape[0]} samples")
print(f"Class distribution (test): {np.bincount(y_test)}")
print("\nSetup complete ✓")
```

## Task 2: Bootstrap Confidence Intervals

For each model, compute **95 % bootstrap confidence intervals** for F1, precision, and recall.

Steps:
1. Write a reusable function `bootstrap_metric(y_true, y_pred, metric_fn, n_boot=2000, seed=42)`.
2. Apply it to each model and metric.
3. Visualize with a grouped bar chart or dot-and-whisker plot (error bars = 95 % CI).
4. Answer the interpretation question below.

```python
def bootstrap_metric(y_true, y_pred, metric_fn, n_boot=2000, seed=42):
    """Compute bootstrap 95% confidence interval for a classification metric.

    Returns
    -------
    point_estimate : float
    ci_lower : float
    ci_upper : float
    boot_scores : np.ndarray
    """
    
    rng = np.random.default_rng(seed)
    n = len(y_true)

    boot_scores = []

    for _ in range(n_boot):
        # Random indekslər seç
        indices = rng.choice(n, size=n, replace=True)

        # Həmin indekslər üzrə sample götür
        y_true_sample = y_true[indices]
        y_pred_sample = y_pred[indices]

        # Metric hesabla
        score = metric_fn(y_true_sample, y_pred_sample)
        boot_scores.append(score)

    boot_scores = np.array(boot_scores)

    # Original metric
    point_estimate = metric_fn(y_true, y_pred)

    # 95% confidence interval
    ci_lower = np.percentile(boot_scores, 2.5)
    ci_upper = np.percentile(boot_scores, 97.5)

    return point_estimate, ci_lower, ci_upper, boot_scores
    
```

```python
# Apply bootstrap_metric to each model for F1, precision, and recall
metrics = {
    "F1": f1_score,
    "Precision": precision_score,
    "Recall": recall_score
}

models = {
    "Logistic Regression": pred_A,
    "Random Forest": pred_B
}

results = []

for model_name, preds in models.items():
    for metric_name, metric_fn in metrics.items():

        point, lower, upper, scores = bootstrap_metric(
            y_test, preds, metric_fn
        )

        results.append({
            "Model": model_name,
            "Metric": metric_name,
            "Score": point,
            "CI Lower": lower,
            "CI Upper": upper
        })

results_df = pd.DataFrame(results)

print("\nBootstrap Results:\n")
results_df
```

## Task 3: Paired Permutation Test

Formally compare the two best-performing models:

1. Record per-sample correctness for each model (1 = correct, 0 = wrong).
2. Run a two-sided permutation test on the paired accuracy differences.
3. Report the observed difference, the p-value, and a plain-language conclusion (α = 0.05).
4. Repeat the test for F1 (bootstrap the F1 difference or use the micro-averaging trick).

```python

correct_A = (pred_A == y_test).astype(int)
correct_B = (pred_B == y_test).astype(int)

print("Correct A (first 10):", correct_A[:10])
print("Correct B (first 10):", correct_B[:10])
```

```python
import numpy as np


correct_A = (pred_A == y_test).astype(int)
correct_B = (pred_B == y_test).astype(int)


diff = correct_B - correct_A

observed_diff = np.mean(diff)

print("Observed accuracy difference:", observed_diff)


n_perm = 2000
rng = np.random.default_rng(42)

perm_diffs = []

for _ in range(n_perm):
    # random sign flip (paired structure preserved)
    signs = rng.choice([-1, 1], size=len(diff))
    
    perm_diff = np.mean(diff * signs)
    perm_diffs.append(perm_diff)

perm_diffs = np.array(perm_diffs)


p_value = np.mean(np.abs(perm_diffs) >= abs(observed_diff))

print("P-value:", p_value)


alpha = 0.05

if p_value < alpha:
    print("Conclusion: Statistically significant difference between models")
else:
    print("Conclusion: No statistically significant difference between models")
```

```python
# Repeat for F1-score
import numpy as np
from sklearn.metrics import f1_score


def bootstrap_f1_diff(y_true, pred_A, pred_B, n_boot=2000, seed=42):
    rng = np.random.default_rng(seed)
    n = len(y_true)

    f1_diffs = []

    for _ in range(n_boot):
        # resample indices (paired bootstrap)
        idx = rng.choice(n, size=n, replace=True)

        y_sample = y_true[idx]
        A_sample = pred_A[idx]
        B_sample = pred_B[idx]

        f1_A = f1_score(y_sample, A_sample)
        f1_B = f1_score(y_sample, B_sample)

        f1_diffs.append(f1_B - f1_A)

    f1_diffs = np.array(f1_diffs)

    # observed difference
    observed = f1_score(y_true, pred_B) - f1_score(y_true, pred_A)

    # confidence interval
    lower = np.percentile(f1_diffs, 2.5)
    upper = np.percentile(f1_diffs, 97.5)

    return observed, lower, upper, f1_diffs



f1_obs, f1_low, f1_up, f1_diffs = bootstrap_f1_diff(
    y_test, pred_A, pred_B
)

print("=== F1-score Bootstrap Test ===")
print("Observed F1 difference:", f1_obs)
print("95% CI:", (f1_low, f1_up))



if f1_low > 0 or f1_up < 0:
    print("Conclusion: Statistically significant difference in F1-score")
else:
    print("Conclusion: No statistically significant difference in F1-score")
```
